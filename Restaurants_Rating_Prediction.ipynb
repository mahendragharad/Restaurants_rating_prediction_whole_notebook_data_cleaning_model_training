```python
# function to describe the dataset 
def describe_dataset(df) :

    unique_values = []
    n_unique_values = []

    col = ['Data_type', 'Missing_value_Count', 'Unique_Values', 'N_unique'] 
    DSC_df = pd.DataFrame(columns=col)
    DSC_df["Missing_value_count"] = df.isnull().sum()
    DSC_df["Data_type"] = df.dtypes

    for i in data :
        unique_values.append(data[i].unique())
        n_unique_values.append(len(data[i].unique()))
    
    DSC_df["Unique_Values"] = unique_values 
    DSC_df["N_unique"] = n_unique_values

    DSC_df.drop("Missing_value_Count" , axis = 1 , inplace = True )

    return DSC_df
```

```python
def find_top_categories(df):
    top_categories = {}
    for column in df:
        
        if df[column].dtype == "O" :
            # Count the frequency of each category in the column
            category_counts = df[column].value_counts()

            # Extract the top 5 categories (or fewer if there are fewer than 5 unique categories)
            top_5_categories = category_counts.head(5).index.tolist()

            # Store the top categories in the dictionary
            top_categories[column] = top_5_categories
    
    return top_categories
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Assuming you have already defined data_copy

# Define a list of the 10 columns you want to encode
columns_to_encode = ['City' , 'Address' , 'Cuisines' , 'Rating color']

threshold = 5  # Number of top categories to keep

encoder = OneHotEncoder(sparse=False , drop= "first")

# Loop through each column
for column in columns_to_encode:
    # Get the value counts for the column
    counts = data_copy[column].sort_values(ascending=False).value_counts()
    
    # Identify the top categories to keep and replace the rest with 'uncommon'
    top_categories = counts.index[:threshold]
    data_copy[column] = data_copy[column].where(data_copy[column].isin(top_categories), 'uncommon')
    
    # Fit and transform the column with one-hot encoding
    encoded_data = encoder.fit_transform(data_copy[[column]])
    
    # Create a DataFrame from the one-hot encoded values
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out([column]))
    
    # Concatenate the one-hot encoded DataFrame with the original data_copy DataFrame
    data_copy = pd.concat([data_copy, encoded_df], axis=1)
    
    # Drop the original column
    data_copy.drop(columns=[column], inplace=True)

# Sample 5 rows of the encoded DataFrame
sampled_data = data_copy.sample(5)

print(sampled_data)
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
# To train multiple models 
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    
    # Here the models getting trained 
    model.fit(X_train,y_train)

    #Make Predictions always based on test data
    y_pred=model.predict(X_test)

    # By obtaining the predicted values, you can compare them 
    # with the actual target values (y_test) to evaluate the model's
    # performance and assess how well it predicts the 
    # target variable on unseen data.

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
